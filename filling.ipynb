```python
import pandas as pd
df = pd.read_csv("week5_categoriconly.csv", nrows = 100000)
df
```

```python
#1
for c in df.columns:
    df[c] = df[c].map( {'Y': 1, 'N': 0, 'U': 0 } )#herfleri 0 ve 1 lerle evez edir
df.fillna(0,inplace=True) #null\nan deyerleri 0 la evez edir 
print(df) 
```

```python
#2
list1=[]
list2=[]
for c in df.columns:
    if df[c].isnull().any():#null deyeri olan columlar
        list1.append(c)
    if df[c].notnull().any():#null deyeri olmyan columlar
        list2.append(c)
print('dfa:',list1,'dfB:',list2)
```

```python
#3
for c in df.columns:
    df[c] = df[c].fillna( df[c].mode()[0] )
    #print(df[c].isnull().sum())
print(df.isna().sum())    
```

```python
df[ df['ownrent'].isnull() ]
```

```python
#4
for c in df.columns:
    type=str(df[c].dtype)
    if type in ['int64', 'float64']:
        df[c] = df[c].fillna( df[c].mean() )
    #print(df[c].isnull().sum())
print(df.select_dtypes(include = 'int64' ).isna().sum())   
```

```python
for c in df.columns:
    type=str(df[c].dtype)
    if type in ['int64', 'float64']:
        df[c] = df[c].fillna( df[c].median() )
    #print(df[c].isnull().sum())
print(df.select_dtypes(include = 'int64' ).isna().sum())  
```

```python
#5
for c in df.columns:
    for d in df.columns:
        type=str(df[c].dtype)
        type1=str(df[d].dtype)
        if type in ['int64', 'float64'] and type1 in ['int64', 'float64']:
            corr=df[c].corr(df[d])
            print(c,d,corr)        
```
